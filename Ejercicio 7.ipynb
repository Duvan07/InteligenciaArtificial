```python
x_n = [1,1]
y_n = [1,2]
x_b = [2,2]
y_b = [-1,0]

x_c = [-1,-2]
y_c = [2,1]
x_d = [-1,-2]
y_d = [-1,-2]

x_f = [-2,2]
y_f = [-1,1]

x_e = [-1,1]
y_e = [3,-3]
plt.title("ejercicio1")
plt.xlabel("Valor X")
plt.ylabel("Valor y")

plt.plot(x_n,y_n, linestyle="none", marker="o", color="red")
plt.plot(x_b,y_b, linestyle="none", marker="x", color="blue")
plt.plot(x_c,y_c, linestyle="none", marker="*", color="green")
plt.plot(x_d,y_d, linestyle="none", marker="+", color="black")
plt.plot(x_f,y_f, color="black")
plt.plot(x_e,y_e, color="black")

plt.grid()
plt.show()
```
